## Modelo de Recomendação com FPGrowth - Cestas

### Principais Conceitos

O FPGrowth é um modelo de Machine Learning voltado para a criação de recomendações baseadas em **cestas**. 
Nesse contexto, uma cesta são itens que possuem uma ação em conjunto. 

O modelo é baseado em **frequência**, portanto, ele tem a capacidade de separar cada item de acordo com sua relação com outros itens e entender sua importância para aquele determinado nó.

Para que seja mais claro o entendimento conceitual, o modelo FPGrowth pode ser comparado a uma árvore. Onde cada ramificação está um 'item' que pode ter relação com outros 'itens'.

Nossos parâmetros são limitantes, escolhendo qual galho deve ser ignorado de acordo com a frequência observada.



### Organização dos Dados

Para que seja possível essa explosão dos dados em ramificações e frequência, o FPGrowth exige um tratamento inicial dos dados para que eles sejam agrupados em um formato determinado.

*Para fins didáticos, usaremos exemplos de comportamentos de compras em um determinado negócio.*

O formato determinado precisa ser uma linha para cada compra e os produtos devem estar agrupados em um *array*.

| 123 | [I1, I2, I5] |

| 126 | [I1, I2, I4] |
| 127 | [I1, I3] |
| 128 | [I2, I3] |
| 129 | [I1, I3] |
| 130 | [I1, I2, I3, I5] |
| 131 | [I1, I2, I3] |
